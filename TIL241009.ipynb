함수의 기본 구조
Python에서 함수를 정의하려면 def 키워드를 사용한다.

함수 이름을 지정하고, 필요한 경우 매개변수를 정의한다.

함수는 필요한 작업을 수행한 후, 반환값을 돌려줄 수 있다.

 

def 함수이름(매개변수1, 매개변수2):
    # 실행할 코드
    return 반환값

 

def greet(name):
    return f"안녕하세요, {name}님!"

message = greet("Alice")
print(message)  # 안녕하세요, Alice님!

 

매개 변수
매개변수는 함수가 호출될 때 입력받는 값(인수, arguments)을 처리하기 위해 사용하는 변수다. 함수에 전달된 인수는 매개변수에 할당되며, 함수 내부에서 사용된다.

 

def add(a, b):
    return a + b

result = add(3, 5)
print(result)  # 8

 

다중 매개 변수
함수는 여러 개의 매개변수를 가질 수 있다. 각 매개변수는 콤마(,)로 구분한다.

 

def multiply(x, y, z):
    return x * y * z

result = multiply(2, 3, 4)
print(result)  # 24

 

반환값
반환값(Return Value)은 함수가 작업을 완료한 후, 호출된 위치로 돌려주는 값이다. return 키워드를 사용하여 반환값을 지정할 수 있다.

 

def square(n):
    return n * n

result = square(5)
print(result)  # 25

 

복수 반환값
Python에서는 함수가 여러 값을 반환할 수 있다. 이때 반환값은 튜플로 반환된다.

 

def calculate(a, b):
    sum = a + b
    difference = a - b
    return sum, difference

s, d = calculate(10, 5)
print(s)  # 15
print(d)  # 5

 

반환값이 없는 함수
함수에서 return 문을 생략하거나 반환값 없이  return만 사용하면, 함수는 None을 반환한다.

 

def say_hello():
    print("Hello!")

result = say_hello()  # None 반환
print(result)  # None

 

기본값 (Default Parameter)
def greet(name="손님"):
    return f"안녕하세요, {name}님!"

print(greet())          # 안녕하세요, 손님님!
print(greet("Alice"))   # 안녕하세요, Alice님!

 

다중 기본값
함수는 여러 매개변수에 대해 기본값을 가질 수 있다. 다만, 기본값이 있는 매개변수는 항상 기본값이 없는 매개변수 뒤에 와야 한다.

 

def order(item, quantity=1, price=1000):
    return f"{item} {quantity}개 주문하셨습니다. 총 가격은 {quantity * price}원입니다."

print(order("사과"))                   # 사과 1개 주문하셨습니다. 총 가격은 1000원입니다.
print(order("바나나", 2))               # 바나나 2개 주문하셨습니다. 총 가격은 2000원입니다.
print(order("딸기", 3, 1500))          # 딸기 3개 주문하셨습니다. 총 가격은 4500원입니다.
