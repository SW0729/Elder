클래스와 객체의 개념
 클래스란 객체를 만들기 위한 청사진(Template) 또는 설계도, 속성과 메서드로 이뤄진다. 속성은 객체의 데이터를, 메서드는 객체의 동작을 정의한다.

 

 객체란 클래스의 인스턴스로, 클래스라는 설계도로부터 실제로 생성된 구체적인 데이터와 기능을 가지는 실체다. 객체는 클래스에 정의된 속성과 메서드를 가진다.

 

class Dog:  # 클래스 정의
    def __init__(self, name, breed):  # 초기화 메서드
        self.name = name  # 속성 정의
        self.breed = breed

    def bark(self):  # 메서드 정의
        return f"{self.name}가 짖습니다."

# 객체 생성
my_dog = Dog("Buddy", "Golden Retriever")
print(my_dog.bark())  # Buddy가 짖습니다.

속성과 메서드 살펴보기
 속성은 객체의 상태나 특성을 나타내는 변수이다. 속성은 클래스 내부에서 정의되며, 각각의 개체는 고유한 속성값을 가질 수 있다. 메서드는 객체가 수행하는 동작을 정의하는 함수다. 메서드는 클래스 내부에 정의되며, 객체의 속성을 조작하거나 특정 동작을 수행할 수 있다.

 

파일 모드
"r": 읽기 모드(기본값). 파일이 존재하지 않으면 오류가 발생한다.
"w": 쓰기 모드. 파일이 존재하지 않으면 새로 만들고, 파일이 존재하면 내용을 덮어쓴다.
"a": 추가 모드. 파일 끝에 데이터를 추가한다.
"b": 바이너리 모드. 이진 파일을 읽거나 쓸 때 사용하며, 다른 모드와 함께 사용한다(예: "rb", "wb").
