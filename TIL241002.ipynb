시리즈(Series)란
시리즈란 1차원 배열과 같은 데이터 구조
엑셀의 한 열(Column)과 유사
인덱스(Index)와 데이터 값(Value)이 쌍으로 구성
import pandas as pd

# 예시: 시리즈 생성
s = pd.Series([10, 20, 30, 40], index=['a', 'b', 'c', 'd'])
print(s)

 

시리즈의 특징

인덱스를 통한 데이터로의 빠른 접근
데이터 타입은 자동으로 설정되지만, 원하는 타입으로 변경 가능
데이터 프레임 이해
2차원의 표 형태 데이터
엑셀의 전체 시트와 유사
여러 개의 시리즈가 모여 만들어짐
# 예시: 데이터 프레임 생성
data = {
    '이름': ['철수', '영희', '민수'],
    '나이': [25, 30, 35],
    '직업': ['학생', '회사원', '프리랜서']
}

df = pd.DataFrame(data)
print(df)

 

데이터 프레임의 특징

행(row)과 열(column)로 구성된 데이터
다양한 데이터 타입을 각 열에 담을 수 있다.
인덱스와 컬럼명으로 특정 데이터에 접근할 수 있다.
데이터 프레임에서 시리즈 다루기
데이터 프레임의 각 열(column)은 시리즈다.
# 특정 열을 시리즈로 선택
age_series = df['나이']
print(age_series)

 

시리즈와 데이터 프레임의 공통점과 차이점
공통점
둘 다 인덱스가 있어 데이터에 쉽게 접근 가능
판다스에서 가장 기본적 데이터 구
차이점
시리즈는 1차원, 데이터 프레임은 2차원
시리즈는 하나의 데이터 타입, 데이터 프레임은 여러 개의 데이터 타입이 허용
 
데이터를 선택하는 기본 방법
인덱스를 사용한 데이터 선택
set_index()로 인덱스 설정하기
특정 열을 인덱스로 설정하면, 데이터 선택이 더 직관적
import pandas as pd

# 예시 데이터프레임 생성
data = {
    '이름': ['철수', '영희', '민수'],
    '나이': [25, 30, 35],
    '직업': ['학생', '회사원', '프리랜서']
}
df = pd.DataFrame(data)

# '이름' 열을 인덱스로 설정
df = df.set_index('이름')
print(df)

 

loc[]을 사용해 인덱스로 데이터 선택
인덱스를 설정하고 loc[]을 사용해 쉽게 데이터에 접근
# '영희'의 데이터 선택
print(df.loc['영희'])

 

열(Column) 지정하여 데이터 선택하기
단일 열 선택
열 이름을 사용해 단일 열을 선택한다.
# '나이' 열 선택
print(df['나이'])

 

여러 열 선택
여러 열을 선택하고 싶다면, 열 이름의 리스트를 사용
# '나이'와 '직업' 열 선택
print(df[['나이', '직업']])

 

행(Row) 지정하여 데이터 선택하기
iloc[]을 사용한 행 선택

iloc[]을 사용해 인덱스 라벨로 특정 행 선택
# '민수'의 데이터 선택
print(df.loc['민수'])
